```python
import csv
import pandas as pd
from pathlib import Path
```

```python
menu_csv = Path("Resources/menu_data.csv")
menu_df = pd.read_csv(menu_csv, index_col="item")
menu_df.head()
```

```python
sales_csv = Path("Resources/sales_data.csv")
sales_df = pd.read_csv(sales_csv, index_col="Menu_Item")
sales_df.head()
```

```python
menu_df = menu_df.drop(columns=['category','description'])
menu_df = menu_df.sort_values('item',ascending=True)
menu_df.head()
```

```python
sales_df = sales_df.drop(columns=['Line_Item_ID', 'Date','Credit_Card_Number'])
sales_df = sales_df.rename(columns={'Menu_Item': 'item'})
sales_df = sales_df.sort_values("Menu_Item", ascending=True)
sales_df = sales_df.groupby("Menu_Item")
```

```python
sales_df.head()
```

```python
sum_df = sales_df['Quantity'].sum()
```

```python
sum_df.head()
```

```python
joined_df = pd.concat([sum_df,menu_df], axis = "columns", join = "inner")
joined_df.head()
```

```python
joined_df['02-revenue'] = joined_df["Quantity"] * joined_df["price"]
joined_df['03-cogs'] = joined_df["Quantity"] * joined_df["cost"]
joined_df['04-profit'] = joined_df['02-revenue'] - joined_df['03-cogs'] 
```

```python
joined_df.drop(columns=['price', 'cost'], inplace=True)
joined_df.head()
```

```python
report_df = joined_df.rename(columns={"Quantity":"01-count"})
report_df.index.name = 'item'
report_df.head()
```

```python
report_df.to_csv("report.csv")
```
